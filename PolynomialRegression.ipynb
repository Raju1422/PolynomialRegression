```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.arange(1,25,0.5)
```

```python
y = 0.5 * x + 1.29 * np.square(x) + 35*(x**5) + np.random.randint(-100, 100, size=len(x))
```

```python
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
X_train,X_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=7)

model = LinearRegression()
model.fit(X_train.reshape(-1,1),y_train.reshape(-1,1))
r2_score(y_test,model.predict(X_test.reshape(-1,1)))
```

```python
plt.scatter(x,y)
plt.plot(x,model.predict(x.reshape(-1,1)),c='r')
plt.legend(['orginal','predicted'])
```

```python
r2_score(y_test,model.predict(X_test.reshape(-1,1)))
```

## Conclusion : From linear regression we get r2 score is 0.427

```python
from sklearn.preprocessing import PolynomialFeatures
pr = PolynomialFeatures(degree=2)
x_2=pr.fit_transform(x.reshape(-1,1))
X_train =pr.fit_transform(X_train.reshape(-1,1))
X_test= pr.transform(X_test.reshape(-1,1))

print(X_train.shape)
y_train.shape
```
